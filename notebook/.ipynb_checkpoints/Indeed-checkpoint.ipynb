```python
import pandas as pd
from matplotlib import pyplot as plt
from indeed import cfg
import seaborn as sns
from indeed.functional import normalize_features
```

```python
xdf, ydf = pd.read_csv(cfg.etl.xtrain_csv), pd.read_csv(cfg.etl.ytrain_csv)
```

```python
xdf = normalize_features(xdf.filter(cfg.features.cat+cfg.features.num))
xdf
```

```python
ydf
```

```python
ydf['salary'].describe()
```

```python
sns.distplot(ydf['salary'])
```

```python
#skewness and kurtosis
print("Skewness: %f" % ydf['salary'].skew())
print("Kurtosis: %f" % ydf['salary'].kurt())
```

```python
for c in xdf.columns:
    print("=======")
    print(xdf[c].describe())
```

```python
def cat_view(var):
    data = pd.concat([ydf['salary'], xdf[var]], axis=1)
    f, ax = plt.subplots(figsize=(8, 6))
    fig = sns.boxplot(x=var, y="salary", data=data)
    fig.set_title(f"corr :{xdf[var].corr(ydf['salary'])}")
```

```python
_ = [cat_view(f) for f in cfg.features.cat]
```

```python
_ = [cat_view(f) for f in cfg.features.num]
```
